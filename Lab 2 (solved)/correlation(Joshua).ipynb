## Часть 3: Вычисление корреляции с Python


#### Шаг 1: Рассчитайте корреляцию для brainFrame.
Метод `corr()` библиотеки pandas обеспечивает простой способ вычисления корреляции для кадра данных. Просто применяя метод на кадре данных, можно получить корреляцию между всеми переменными одновременно.

```python
correlation_matrix = brainFrame.select_dtypes(include=['number']).corr(method='pearson')

print(correlation_matrix.round(2))
```

Обратите внимание на диагональ слева направо в таблице корреляции, сгенерированной выше. Почему диагональ заполнена значениями 1? Это совпадение? Объясните.

Диагональ показывает 1.0, потому что это корреляция каждой переменной с самой собой.
- Когда вы сравниваете любую величину саму с собой (например, FSIQ с FSIQ) - это идеальное совпадение
- На графике рассеяния все точки выстроятся в идеальную линию
- Это даёт максимально возможное значение корреляции - 1.0

Это не случайность - так всегда происходит в корреляционных матрицах:
- Диагональ всегда будет 1.0 при правильных расчётах
- Это как смотреть в зеркало - вы всегда идеально совпадаете с самим собой

-- Пример:
Если Рост коррелирует сам с собой:
`Рост vs Рост = 1.0` (всегда)

Продолжая смотреть на таблицу корреляции выше, обратите внимание, что значения зеркалируются; значения под диагональю имеют зеркальный аналог над ней. Это совпадение? Объясните.

-- Oтвет:
Нет, это не совпадение. Так и должно быть!

-- Oбъяснение:

1. Корреляция между A и B = корреляции между B и A.

- Пример: связь роста с весом = связи веса с ростом.

2. Матрица просто "зеркалит" одинаковые значения над и под диагональю.

-- Почему?

- Корреляция измеряет взаимную связь двух переменных.

- Порядок переменных не влияет на результат.

-- Пример:

          Рост  Вес
Рост      1.0  0.6  
Вес       0.6  1.0  

0.6 над диагональю = 0.6 под диагональю — это одна и та же связь.

-- Вывод:
Это особенность корреляционных матриц, а не ошибка. Так работает математика!

#### Шаг 2: Нарисуйте корреляционную тепловую карту.

Теперь, когда кадры данных готовы, можно отобразить тепловые карты. Далее приведен анализ кода из ячейки ниже:

Строка 1: Создает таблицу корреляции, основанную на фрейме данных `womenNoGenderDf` и хранит его в `wcorr`.<br>
Строка 2: Использует метод `heatmap()` библиотеки `seaborn` для генерации и построения тепловой карты. Обратите внимание, что `heatmap()` принимает `wcorr`, как параметр.<br>
Строка 3: используется для экспорта и сохранения созданной тепловой карты в виде PNG-изображения. Хотя строка 3 не активна (перед ней стоит символ `#`, заставляя интерпретатор игнорировать ее), она сохранена в информационных целях.

```python
# Create correlation matrix for women
wcorr = womenDf.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))

# Create heatmap with better formatting
heatmap = sns.heatmap(
    wcorr, 
    annot=True,            # Show correlation values in cells
    fmt=".2f",            # Format to 2 decimal places
    cmap="coolwarm",      # Blue-red color scheme
    vmin=-1, vmax=1,      # Fix color scale from -1 to 1
    linewidths=0.5,       # Add lines between cells
    square=True           # Keep cells square
)

plt.title('Correlation Matrix (Women)', pad=20)


plt.show()
```

Аналогичным образом создайте и отобразите тепловую карту для кадра данных с мужчинами.

```python
# For male data (same structure)
mcorr = menDf.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    mcorr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5
)
plt.title('Correlation Matrix (Men)', pad=20)
plt.show()
```

У многих пар переменных корреляция близка к нулю. Что это значит?

-- Oтвет:
Если корреляция близка к нулю — это значит, что между этими переменными нет линейной связи.

-- Пример из данных:
Если корреляция между IQ и ростом ≈ 0, это означает, что:
- Высокие люди не обязательно умнее
- Умные люди не обязательно выше

-- Важно:
Нулевая корреляция — это не значит, что связи нет вообще. Возможны:

1. Нелинейная связь (например, U-образная)

2. Связь есть, но только в подгруппах данных

-- Вывод:
Корреляция ≈ 0 → искать другие типы зависимостей или признать переменные независимыми.

Зачем делать разделение по полу?

1. Физиологические различия

- У мужчин и женщин могут быть разные средние значения (например, размер мозга, рост, вес).

- Если не разделить данные, корреляция может исказиться из-за этих различий.

2. Разная структура взаимосвязей

- Например, связь между IQ и размером мозга может быть сильнее у мужчин, чем у женщин (или наоборот).

- Если смешать данные, можно пропустить важные закономерности.

3. Избегаем "ложных корреляций"

- Если в данных у мужчин в среднем больше и мозг, и IQ, а у женщин — меньше, общая корреляция может казаться сильной, даже если внутри каждой группы связи нет.

Пример из ваших данных

- Допустим, у мужчин в среднем MRI_Count (размер мозга) больше, чем у женщин.

- Если не разделять данные, может показаться, что больший мозг = выше IQ, даже если на самом деле это просто разница между полами.

Разделение по полу помогает:
- Увидеть реальные взаимосвязи внутри каждой группы.
